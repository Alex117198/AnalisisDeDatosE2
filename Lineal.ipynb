```python
import matplotlib.pyplot as plt
from sklearn import linear_model
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
#definir data con la ruta del archivo
data =  pd.read_csv("C:/Users/ALYE1/OneDrive/Documents/La U mi perro/S9/An√°lisis de Datos/T2/lineal/data.csv", sep=",")
```

```python
#mostrar datos
data
```

```python
#definir variables
x = data[["metro"]]
y = data["precio"]
```

```python
#mostrar la grafica
data.plot.scatter(x="metro", y="precio")
```

```python
#crear el modelo de regreci√≥n lineal
lrModel = linear_model.LinearRegression()
lrModel.fit(x, y)
```

```python
#realizar prediccion en x
prediction = lrModel.predict(x)
```

```python
#imprimir intersecci√≥n y pendiente
print("Intersecci√≥n (b): ", lrModel.intercept_)
print("Pendiente (m): ", lrModel.coef_)
```

```python
#mostrar resultados
plt.scatter(x, y, label= "Datos reales")
plt.plot(x, prediction, color="red", label="Regresi√≥n Lineal")
plt.xlabel("Metros")
plt.ylabel("Precio")
plt.legend()
plt.show()
```
